```python
import json

import pickle
import numpy as np
```

```python
def load_data_json(json_path):
    f = open(json_path, 'r', encoding='utf-8')
    json_data = json.load(f)
    f.close()
    return json_data
```

```python
mode = 'public'
json_path = f'./data/{mode}_complete.json'
json_data = load_data_json(json_path)

results_path = f'./output_results/public_result_colbert30'
with open(results_path, 'rb') as handle:
    results = pickle.load(handle)
            
```

```python
print(len(json_data))
```

```python
did2data = {}
for d in json_data:
    did = d['did']
    did2data[did] = d
```

```python
topk = 20

pos_pairs = []

for k, values in results.items():
    did = str(k)
    
    data = did2data[did]
    pos_dids = data['pos_dids']
    if mode=='train':
        if len(pos_dids) == 0:
            continue
        
    for pos_did in pos_dids:
        train_binary_pairs.append((did, pos_did, 1))
        train_pos_count += 1
    print(f'did = {k}')
    
    bm25_pos_dids = data['bm25_pos_dids'][:topk]
    
    all_ref_ids = np.array([v[0] for v in values])
    scores = [v[2] for v in values]
    scores = np.array(scores)
    
    
    mean = np.mean(scores)
    std = np.std(scores)
    max_score = np.max(scores)
    
    print(f' mean {mean}, std {std}, max {max_score}')

    th = mean + 2 * std
    pos_idx = scores >= th

    
    
    print('score based predict topk did and score')
    pos_ref_ids = list(all_ref_ids[pos_idx])
    print(pos_ref_ids)
    
    for pos_ref_id in pos_ref_ids:
        pos_pairs.append((did, pos_ref_id))

    
```
